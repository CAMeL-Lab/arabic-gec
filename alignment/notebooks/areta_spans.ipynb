```python
import copy
```

```python
def read_data(path):
    with open(path) as f:
        return [x.strip() for x in f.readlines()]
```

```python
def add_spans_areta_tags(path):
    example = []
    all_examples = []
    
    with open(path, mode='r') as f:
        for line in f.readlines():
            line = line.strip().split('\t')
            if len(line) > 1:
                assert len(line) == 3
                
                example.append({'src_token':line[0], 'tgt_token': line[1],
                                'tag': line[2]})
            else:
                all_examples.append(example)
                example = []
    
    return all_examples
```

```python
def add_spans(examples, src_sents):
    curr_idx = 0
    new_examples = []
    
    assert len(examples) == len(src_sents)

    for i, example in enumerate(examples):
        start, end = 0, 0
        
        src_sent = src_sents[i].split()
        new_example = copy.copy(example)
    
        for ex in new_example:
            end = start + len(ex['src_token'].split())
            ex['span'] = (start, end)
            
            assert " ".join(src_sent[start:end]) == ex['src_token']

            start = end
            
        new_examples.append(new_example)
    
    return new_example
```
